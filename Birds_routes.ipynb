```python
# функция для поиска птиц, которые пролетают на расстоянии 200 км от выбранной точки

def find_route(name):
    # read file
    res = []
    new_res = []
    for i in range(0, len(name)):
        print('Processing ', i+1, 'out of', len(name), ':', name[i])
        file_name = '/content/drive/MyDrive/Птицы/' + name[i]
        df = pd.read_csv(file_name)
        df = df[['timestamp', 'location_long', 'location_lat', ]]
        df['timestamp']= pd.to_datetime(df['timestamp'])
        df = df.loc[(df['timestamp'] >= '2021-03-01') & (df['timestamp'] <= '2021-05-31')]
        df['timestamp'] = pd.to_datetime(df['timestamp']).dt.date
    
    # выбираем птиц, которые пролетают ближе 201 ки от точки
        
        for j in range(0, len(df)):
            coords_1 = north_point_coords
            try:
                coords_2 = (df['location_lat'].iloc[j], df['location_long'].iloc[j])
            
                if geopy.distance.geodesic(coords_1, coords_2).km < 201:
                    res.append(name[i])
                    print('Result: Yes')
                    break
                                    
            except:
                print('Error')
                continue
        
    print('Finished, results:', len(res))
    # в результате выдает список искомых файлов
    return res
```

```python
# функция для выбора координат для графика
def coordinates(name):
    coordinates = []
    
    for i in range(0, len(name)):
        
        lats = []
        lons = []
        ids = []
        label = ''
        
        print('Processing ', i+1, 'out of', len(name), ':', name[i])
        file_name = '/content/drive/MyDrive/Птицы/' + name[i]
        
        df = pd.read_csv(file_name)
        df = df[['timestamp', 'location_long', 'location_lat', ]]
        df['timestamp']= pd.to_datetime(df['timestamp'])
        df = df.loc[(df['timestamp'] >= '2021-03-01') & (df['timestamp'] <= '2021-05-31')]
        df['timestamp'] = pd.to_datetime(df['timestamp']).dt.date
        
        lons = df['location_long'].tolist()[::100]
        lats = df['location_lat'].tolist()[::100]
        ids = df['timestamp'].tolist()[::100]
        label = name[i]
        
        coordinates.append([lons, lats, ids, label])
        
    return coordinates
        
```
